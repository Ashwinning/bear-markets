```python
import json

with open('layoffs-fyi.json', 'r') as f:
    response = json.load(f)
```

```python
import pandas as pd
```

```python
rows = [x['cellValuesByColumnId'] for x in response['data']['table']['rows']]
```

```python
newDict = []
for row in rows:
    entry = {}
    for key in row.keys():
        name = [r['name'] for r in response['data']['table']['columns'] if r['id'] == key][0]
        rowType = [r['type'] for r in response['data']['table']['columns'] if r['id'] == key][0]
        if rowType in ['text', 'number', 'date']:
            entry[name] = row[key]
        if rowType in ['multiSelect']:
            newArr = []
            for x in row[key]:
                newArr.append([r for r in response['data']['table']['columns'] if r['id'] == key][0]['typeOptions']['choices'][x]['name'])
            entry[name] = newArr
    newDict.append(entry)
newDict
```

```python
us_companies = list(filter(lambda x: 'Non-U.S.' not in x['Location'], newDict))
```

```python
df = pd.DataFrame(us_companies)
```

```python
df[df['# Laid Off'] >= 100].to_csv('layoffs-fyi-filtered.csv')
```
