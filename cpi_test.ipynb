```python
import pandas as pd
import plotly.graph_objects as go
```

```python
cpi_df = pd.read_csv('CPIAUCSL.csv')
```

```python
cpi_df['datetime'] = pd.to_datetime(cpi_df.DATE, format='%Y-%m-%d')
cpi_df
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=cpi_df['datetime'], y=cpi_df['CPIAUCSL'], mode='lines', name='Current Market', line={'width': 3, 'color': 'darkblue'}, 
    #hovertemplate=[f" {x['change']:.1%} ({x['timestamp'].strftime('%b %d, %Y')})" for i, x in currentMarket.iterrows()]
    ))
fig.update_layout( 
    yaxis=dict(tickformat="0.2%"),
    xaxis_title = "Trading days since last peak ðŸ †",
    yaxis_title = "ðŸ „ Drawdown",
    template='simple_white',
    paper_bgcolor="rgb(255,255,255)", plot_bgcolor="rgb(255,255,255)",
    legend_traceorder="reversed",
    hovermode='x unified',
    margin=dict(t=50),
    width=1200
)
fig.show()
```
